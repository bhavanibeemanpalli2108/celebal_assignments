#### Prepare Features for Clustering

```python
from sklearn.preprocessing import StandardScaler

# Features for Clustering

cluster_features = [
    'Income',
    'Age',
    'Recency',
    'Total_Spending',
    'Total_Purchases',
    'NumWebVisitsMonth',
    'Customer_Tenure'
]

X = df[cluster_features]

# Standardization

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print("Original Shape:", X.shape)
print("Scaled Shape:", X_scaled.shape)
```

#### Knee Locator Method

```python
# # Run once if needed
# !pip install kneed
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator

inertia = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Knee Detection

knee_locator = KneeLocator(
    k_values,
    inertia,
    curve='convex',
    direction='decreasing'
)

optimal_k = knee_locator.elbow

print("Optimal K:", optimal_k)

# Plot

plt.figure(figsize=(8,5))

plt.plot(k_values, inertia, marker='o')

plt.vlines(
    optimal_k,
    plt.ylim()[0],
    plt.ylim()[1],
    linestyles='dashed',
    colors='red',
    label=f'Optimal K = {optimal_k}'
)

plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Knee Locator for Optimal K")
plt.legend()
plt.grid(True)

plt.show()
```

- Insight: The Knee Locator algorithm was used to objectively identify the optimal number of customer segments by detecting the point of maximum curvature in the inertia curve. This approach reduces subjective interpretation and provides a data-driven estimate of the appropriate cluster count for customer segmentation.

#### Silhouette Analysis

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silhouette_scores = []

k_values = range(2, 11)

for k in k_values:
    
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    
    labels = kmeans.fit_predict(X_scaled)
    
    score = silhouette_score(
        X_scaled,
        labels
    )
    
    silhouette_scores.append(score)

# Results

silhouette_df = pd.DataFrame({
    'K': list(k_values),
    'Silhouette Score': silhouette_scores
})

display(silhouette_df)

# Plot

plt.figure(figsize=(8,5))

plt.plot(
    k_values,
    silhouette_scores,
    marker='o'
)

plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Analysis")
plt.grid(True)

plt.show()

print(
    "Best K based on Silhouette Score:",
    silhouette_df.loc[
        silhouette_df['Silhouette Score'].idxmax(),
        'K'
    ]
)
```

- Insight: Silhouette analysis was performed to evaluate cluster cohesion and separation across different values of K. The optimal cluster count was selected by balancing the KneeLocator recommendation and the highest silhouette score, ensuring both compact and well-separated customer segments

- Overview: While the highest silhouette score was achieved at K=2, this solution produced overly broad customer groups with limited business interpretability. The KneeLocator method identified K=6 as the optimal balance between cluster compactness and segmentation granularity. Therefore, K=6 was selected to generate more actionable customer segments for marketing and customer intelligence applications.

#### Preparing Data for Classification

```python
# Features

X_cls = df.drop(
    columns=[
        'Response',
        'Cluster',
        'Cluster_Name',
        'ID',
        'Dt_Customer'
    ]
)

# Target

y_cls = df['Response']

print("Feature Shape:", X_cls.shape)
print("Target Shape:", y_cls.shape)

print("\nTarget Distribution:")
print(y_cls.value_counts())
```

```python
# encoding categorical values
from sklearn.preprocessing import LabelEncoder

# Create copy

X_cls = X_cls.copy()

# Encode categorical columns

categorical_cols = ['Education', 'Marital_Status']

le = LabelEncoder()

for col in categorical_cols:
    X_cls[col] = le.fit_transform(X_cls[col])

print("Encoding Completed")

X_cls.head()
```

```python
# splitting the dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_cls,
    y_cls,
    test_size=0.2,
    random_state=42,
    stratify=y_cls
)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)

print("\nTraining Target Distribution:")
print(y_train.value_counts(normalize=True))
```

#### feature scaling for distance based models

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling Completed")
```

```python
# evaluation function
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)

results = []

def evaluate_model(model_name, y_true, y_pred, y_prob):
    
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_prob)

    results.append([
        model_name,
        accuracy,
        precision,
        recall,
        f1,
        roc_auc
    ])

    print(f"\n===== {model_name} =====")
    print("Accuracy :", round(accuracy,4))
    print("Precision:", round(precision,4))
    print("Recall   :", round(recall,4))
    print("F1 Score :", round(f1,4))
    print("ROC AUC  :", round(roc_auc,4))

    print("\nClassification Report")
    print(classification_report(y_true, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(
    max_iter=1000,
    random_state=42
)

lr.fit(
    X_train_scaled,
    y_train
)

y_pred = lr.predict(X_test_scaled)

y_prob = lr.predict_proba(
    X_test_scaled
)[:,1]

evaluate_model(
    "Logistic Regression",
    y_test,
    y_pred,
    y_prob
)
```

```python
# Gaussian Naive bayes
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()

nb.fit(
    X_train_scaled,
    y_train
)

y_pred = nb.predict(X_test_scaled)

y_prob = nb.predict_proba(
    X_test_scaled
)[:,1]

evaluate_model(
    "Naive Bayes",
    y_test,
    y_pred,
    y_prob
)
```
